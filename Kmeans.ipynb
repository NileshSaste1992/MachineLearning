# K-Means Clustering Example

```python
import numpy as np
```

```python
##Create fake income/age clusters for N people in k clusters
def createData(n,k):
    np.random.seed(15)
    pointspercluster = float(n)/k
    x = []
    for cluster in range(k):
        income = np.random.uniform(20000.0,200000.0)
        age = np.random.uniform(20.0,70.0)
        for points in range(int(pointspercluster)):
            x.append([np.random.normal(income,10000.0),np.random.uniform(age,2.0)])
    x = np.array(x)
    return x
```

```python
data = createData(100, 5)
```

```python
#Let's see how our data looks
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.figure(figsize=(8,6))
plt.xlabel("Income")
plt.ylabel("Age")
plt.scatter(data[:,0],data[:,1])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
model = KMeans(n_clusters=5)
# scale the data to normalize before giving to model. Will produce better results.
model.fit(scale(data))
model.labels_
```

```python
#To visulalize generated clusters, use colormap which is related to clusters.
plt.figure(figsize=(8,6))
plt.xlabel("Income")
plt.ylabel("Age")
plt.scatter(data[:,0],data[:,1],c=model.labels_.astype(np.float))
plt.show()
```
